````python
from anthropic import Anthropic
from dotenv import load_dotenv
load_dotenv()
client = Anthropic()
````

````python
def add_user_message(messages, prompt):
    messages.append({"role": "user", "content": prompt})

def add_assistant_message(messages, content):
    messages.append({"role": "assistant", "content": content})

def chat(messages, **kwargs):
    with client.messages.stream(        # ← add the 's'
        model="claude-sonnet-4-5",
        max_tokens=1000,
        temperature=0.7,
        messages=messages,              # ← this was missing
        stop_sequences=kwargs.get("stop_sequences", None)
    ) as stream:
        for text in stream.text_stream:
            print(text, end="", flush=True)
        final = stream.get_final_message()

    print()
    assistant_text = final.content[0].text
    add_assistant_message(messages, assistant_text)
    return assistant_text
    
````

````python
messages = []

prompt = """
generate three different sample AWS CLI commands . each should be very short ."""

add_user_message(messages,prompt)
add_assistant_message(messages,"here is the all three comments without any comment :\n```bash")
text = chat(messages, stop_sequences=["```"])
text.strip()
````
