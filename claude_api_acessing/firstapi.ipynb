```python
from dotenv import load_dotenv
load_dotenv()

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
message = client.messages.create(
    model= model,
    max_tokens=1000,
    messages=[{
        "role":"user",
        "content":"tell me about anthropic in one sentence"
    }
    ]

)
```
